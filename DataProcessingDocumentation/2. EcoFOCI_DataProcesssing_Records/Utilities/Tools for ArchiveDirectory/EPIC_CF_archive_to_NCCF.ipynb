# EPIC Archive (with CF time) to NCCF Archive

In the event data needs to be reprocessed from an old completed mooring or cruise, it would be best to convert all the files to the 2020+ CF convention.

These means instead of *.cf.nc files converted from old EPIC files, we will make *.nc files in the final_data_cf folder from the *.cf.nc files in the variable names only:
- lat -> latitude
- lon -> longitude
  

```python
epic2cf_fluor = {
    'lat':'latitude',
    'lon':'longitude',
    'fluor_3031':'fluorometer_volts',
    'Fch_906':'chlor_fluorescence',
}
    
epic2cf_seacat = {
    'lat':'latitude',
    'lon':'longitude',
    'S_41':'salinity',
    'T_20':'temperature',
    'P_1':'pressure',
    'C_50':'conductivity',
}

epic2cf_s37 = {
    'lat':'latitude',
    'lon':'longitude',    
    'S_41':'salinity',
    'T_20':'temperature',
    'P_1':'pressure',
}

epic2cf_s39 = {
    'lat':'latitude',
    'lon':'longitude',    
    'T_20':'temperature',
    'P_1':'pressure',
}
```

```python
import xarray as xa
import numpy as np
```

```python
path = '/Users/bell/ecoraid/2018/Moorings/18bs4b/erddap/final_data/'
name = '18bs4b_s39_0045m.cf.nc'

xdf = xa.load_dataset(f'{path}{name}')
xdf = xdf.astype(np.float64)
```

```python
xdf = xdf.rename(epic2cf_s39)
xdf
```
