```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
import pandas as pd
import xarray as xr
import datetime
import EcoFOCIpy.io.erddap as erddap
```

```python
timeseries ={'url':"http://ecofoci-field.pmel.noaa.gov:8080/erddap",
               'mooringid':'16bs4b',
               'qclevel':'final',
               'instrid':'16bs4b_s37_0053m'
              }
ts1 = erddap.erddapMooredInstretrieve(**timeseries)

ts1 = ts1.set_index(pd.DatetimeIndex(ts1['time (UTC)']))

try:
    ts1.rename(columns={'salinity':'salinity (PSU)'},inplace=True) 
except:
    pass
```

```python
fig = go.Figure()
line = go.Scatter(x=ts1.index.values,y=ts1['salinity (PSU)'].values,mode='markers+lines',)
fig.add_trace(line)
```
