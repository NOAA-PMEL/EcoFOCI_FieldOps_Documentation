# Find CTD and retrieve

using a csv or excel pulled from findclosest to retrieve data

```python
import EcoFOCIpy.io.erddap as erddap

import pandas as pd
```

```python
needed = ['PC01', 'BS10', 'PM02', 'BT01', 'PM04', 'PM05', 'BS11',
       'PM08', 'PM14', 'NM01', 'PH01', 'IC01', 'WT01', 'PB01', 'IC02',
       'BF02', 'WT02', 'IC03', 'HS01', 'HS02']
```

```python
for site in needed:
    edf = pd.read_excel('/Users/bell/Downloads/CTDinfo.xlsx',site,header=None)
    n_df_final = pd.DataFrame()
    n_df_prelim = pd.DataFrame()
    for _i,cast in edf.iterrows():
        print(f'{cast[4]}')
        prelim = {'url':"http://ecofoci-field.pmel.noaa.gov:8082/erddap",
               'cruiseid':cast[4].lower(),
               'qclevel':'preliminary',
               'concastno':cast[1].split('CTD')[1]
              }
        final = {'url':"http://ecofoci-field.pmel.noaa.gov:8080/erddap",
               'cruiseid':cast[4].lower(),
               'qclevel':'final',
               'concastno':cast[1].split('CTD')[1]
              }
        try:
            print(f'trying {site} final')
            df = erddap.erddapCTDretrieve(**final)
            df.columns = [x.split()[0] for x in df.columns]
            n_df_final = pd.concat([n_df_final,
                              df[['time','pressure','latitude',
                                  'longitude','salinity_ch1',
                                  'temperature_ch1','profile_id']]])
            print(f'success {site} final')
            continue
        except Exception as e: 
            print(str(e))
        try:
            print(f'trying {site} prelim')
            df = erddap.erddapCTDretrieve(**prelim)
            df.columns = [x.split()[0] for x in df.columns]
            n_df_prelim = pd.concat([n_df_prelim,
                              df[['time','pressure','latitude',
                                  'longitude','S_41',
                                  'T_28','profile_id']]])
            print(f'success {site} prelim')
        except Exception as e: 
            print(str(e))
        # print(df[['time (UTC)','pressure (dbar)','latitude (degrees_north)','longitude (degrees_east)','salinity_ch1 (PSU)','temperature_ch1 (degree_C)','profile_id']])
        # print(df)
    n_df_final.to_csv(f'{site}.final.csv',index=False)
    n_df_prelim.to_csv(f'{site}.prelim.csv',index=False)
```
