```python
from erddapy import ERDDAP
import matplotlib.pyplot as plt

import pandas as pd
```

```python
server_url = 'http://ecofoci-field.pmel.noaa.gov:8082/erddap'
dataset_name = 'CTD_nw2401_preliminary'
e = ERDDAP(server=server_url,
          protocol='tabledap',
          response='csvp')

e.dataset_id=dataset_name

SBE911_df = pd.read_csv('/Users/bell/ecoraid/2024/CTDcasts/nw2401/nw2401.csv')
SBE911_df.columns = [x.split()[0] for x in SBE911_df.columns]
```

```python
SBE911_origcal_df = pd.read_csv('/Users/bell/ecoraid/2024/CTDcasts/nw2401/nw2401-orig.csv')
SBE911_origcal_df
```

```python
SBE49_df = pd.read_csv('/Users/bell/ecoraid/2024/CTDcasts/nw2401/raw/BONGO/NW2401current/7407_fastcat_data_EcoDAAT_ready_2024-09-19.csv')
SBE49_df_mean = SBE49_df.groupby('Pressure').mean(numeric_only=True)
```

```python
SBE911_df_mean = SBE911_df.groupby('Pressure').mean(numeric_only=True)
```

```python
fig, ax = plt.subplots(2,figsize=(4, 12),sharey=True)
ax[0].plot(SBE911_df_mean.t090C, SBE911_df_mean.index)
ax[0].plot(SBE911_df_mean.t190C, SBE911_df_mean.index)
ax[0].plot(SBE49_df_mean.Temperature, SBE49_df_mean.index)
ax[0].set_ylim([65,0])


ax[1].plot(SBE911_df_mean.sal00, SBE911_df_mean.index)
ax[1].plot(SBE911_df_mean.sal11, SBE911_df_mean.index)
ax[1].plot(SBE49_df_mean.Salinity, SBE49_df_mean.index)
ax[1].set_ylim([65,0])
```

```python
# go profile by profile
SBE49_df.groupby('Station').get_group(1)
```

```python
fig, ax = plt.subplots(2,figsize=(4, 12),sharey=True)
ax[0].plot(SBE911_df.groupby('profileid').get_group('nw240101').t090C,
           SBE911_origcal_df.groupby('profileid').get_group('nw240101')['Pressure [dbar]'])
ax[0].plot(SBE911_df.groupby('profileid').get_group('nw240101').t090C, 
           SBE911_origcal_df.groupby('profileid').get_group('nw240101')['Pressure [dbar]'])
ax[0].plot(SBE49_df.groupby('Station').get_group(1).Temperature, SBE49_df.groupby('Station').get_group(1).Pressure)
ax[0].plot(SBE911_origcal_df.groupby('profileid').get_group('nw240101').t090C, 
           SBE911_origcal_df.groupby('profileid').get_group('nw240101')['Pressure [dbar]'],'--',c='k')
ax[0].plot(SBE911_origcal_df.groupby('profileid').get_group('nw240101').t190C, 
           SBE911_origcal_df.groupby('profileid').get_group('nw240101')['Pressure [dbar]'],'--',c='k')
ax[0].set_ylim([65,0])


ax[1].plot(SBE911_df.groupby('profileid').get_group('nw240101').sal00, 
           SBE911_origcal_df.groupby('profileid').get_group('nw240101')['Pressure [dbar]'])
ax[1].plot(SBE911_df.groupby('profileid').get_group('nw240101').sal11, 
           SBE911_origcal_df.groupby('profileid').get_group('nw240101')['Pressure [dbar]'])
ax[1].plot(SBE49_df.groupby('Station').get_group(1).Salinity, SBE49_df.groupby('Station').get_group(1).Pressure)
ax[1].plot(SBE911_origcal_df.groupby('profileid').get_group('nw240101').sal00, SBE911_origcal_df.groupby('profileid').get_group('nw240101')['Pressure [dbar]'],'--',c='C0')
ax[1].plot(SBE911_origcal_df.groupby('profileid').get_group('nw240101').sal11, SBE911_origcal_df.groupby('profileid').get_group('nw240101')['Pressure [dbar]'],'--',c='C1')
ax[1].set_ylim([65,0])
```

```python
for i in range(1,46):
    fig, ax = plt.subplots(ncols=2,figsize=(8, 4),sharey=True)
    try:
        ax[0].plot(SBE911_df.groupby('profileid').get_group(f'nw2401{str(i).zfill(2)}').t090C, 
                   SBE911_df.groupby('profileid').get_group(f'nw2401{str(i).zfill(2)}')['Pressure'],c='C0')
        ax[0].plot(SBE911_df.groupby('profileid').get_group(f'nw2401{str(i).zfill(2)}').t190C, 
                   SBE911_df.groupby('profileid').get_group(f'nw2401{str(i).zfill(2)}')['Pressure'],c='C1')
    except:
        pass
    try:
        ax[0].plot(SBE49_df.groupby('Station').get_group(i).Temperature, 
                   SBE49_df.groupby('Station').get_group(i).Pressure,c='C3')
    except:
        pass
    ax[0].set_ylim([65,0])
    
    try:
        ax[1].plot(SBE911_df.groupby('profileid').get_group(f'nw2401{str(i).zfill(2)}').sal00, 
                   SBE911_df.groupby('profileid').get_group(f'nw2401{str(i).zfill(2)}')['Pressure'],c='C0')
        ax[1].plot(SBE911_df.groupby('profileid').get_group(f'nw2401{str(i).zfill(2)}').sal11, 
                   SBE911_df.groupby('profileid').get_group(f'nw2401{str(i).zfill(2)}')['Pressure'],c='C1',
                   label=f'ctd{str(i).zfill(3)}')
        ax[1].plot(SBE911_origcal_df.groupby('profileid').get_group(f'nw2401{str(i).zfill(2)}').sal00, 
                   SBE911_origcal_df.groupby('profileid').get_group(f'nw2401{str(i).zfill(2)}')['Pressure [dbar]'],'--',c='C0')
        ax[1].plot(SBE911_origcal_df.groupby('profileid').get_group(f'nw2401{str(i).zfill(2)}').sal11, 
                   SBE911_origcal_df.groupby('profileid').get_group(f'nw2401{str(i).zfill(2)}')['Pressure [dbar]'],'--',c='C1')

    except:
        pass
    try:
        ax[1].plot(SBE49_df.groupby('Station').get_group(i).Salinity, SBE49_df.groupby('Station').get_group(i).Pressure,c='C3')
    except:
        pass
    fig.legend()
    ax[1].set_ylim([65,0])
```
